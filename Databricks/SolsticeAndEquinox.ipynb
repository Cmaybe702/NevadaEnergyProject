```
#Read cleaned weather data 2010

weather_df10 = spark.read.table("weather_annual_2010")

#Add day and year column

from pyspark.sql.functions import substring
from pyspark.sql.functions import to_timestamp
from pyspark.sql.types import TimestampType

weather_df10 = weather_df10.select("minutes", "DateTime", substring("DateTime", 0, 5).alias("Date"), "GlobalHoriz")

weather_df10_c = weather_df10.select("Date", substring("DateTime", 7, 4).alias("Year"), "minutes", "GlobalHoriz")

display(weather_df10_c)
```

```
#Combine data from 2010-2014 and filter solstice and equinox dates

from pyspark.sql.types import IntegerType
from pyspark.sql.functions import col

all_years_df = weather_df14_c.union(weather_df13_c)

all_years_df1 = all_years_df.union(weather_df12_c)

all_years_df2 = all_years_df1.union(weather_df11_c)

all_years_df3 = all_years_df2.union(weather_df10_c)

all_years_df4 = all_years_df3.filter((all_years_df.Date == "06/21") | (all_years_df.Date == "12/21") | (all_years_df.Date == "03/20")
                                   | (all_years_df.Date == "09/22"))

display(all_years_df4)
```

```
#Group by date and time for AVG

all_years_df5 = all_years_df4.groupBy("Date", "minutes").avg()

all_years_df6 = all_years_df5.orderBy("Date")

all_years_df7 = all_years_df6.select("Date", "minutes", col("avg(GlobalHoriz)").alias("AvgGlobalHoriz"))

display(all_years_df7)
```

```
#Group data for visualization

summer_sol = all_years_df7.filter(all_years_df7.Date == "06/21")

summer_sol2 = summer_sol.select("minutes", "AvgGlobalHoriz")

summer_sol2 = summer_sol2.withColumnRenamed("AvgGlobalHoriz", "AvgGlobalHorizSummerSol")

display(summer_sol2)
```

```
winter_sol = all_years_df7.filter(all_years_df7.Date == "12/21")

winter_sol2 = winter_sol.select("minutes", "AvgGlobalHoriz")

winter_sol2 = winter_sol2.withColumnRenamed("AvgGlobalHoriz", "AvgGlobalHorizWinterSol")

display(winter_sol2)
```

```
spring_eq = all_years_df7.filter(all_years_df7.Date == "03/20")

spring_eq2 = spring_eq.select("minutes", "AvgGlobalHoriz")

spring_eq2 = spring_eq2.withColumnRenamed("AvgGlobalHoriz", "AvgGlobalHorizSpringEq")

display(spring_eq2)
```

```
autumn_eq = all_years_df7.filter(all_years_df7.Date == "09/22")

autumn_eq2 = autumn_eq.select("minutes", "AvgGlobalHoriz")

autumn_eq2 = autumn_eq2.withColumnRenamed("AvgGlobalHoriz", "AvgGlobalHorizAutumnEq")

display(autumn_eq2)
```

```
#Categorical Equinox/Solstice data

from pyspark.sql.functions import col, udf
from pyspark.sql.types import StringType

convertUDF = udf(lambda z: convert(z), StringType())

def convert(string):
    if string == "03/20":
        new = "Spring Equinox"
    elif string == "09/22":
        new = "Autumn Equinox"
    elif string == "06/21":
        new = "Summer Solstice"
    elif string == "12/21":
        new = "Winter Solstice"
    return new

all_years_df8 = all_years_df7.select(convertUDF(col("Date").alias("Event")), "minutes", "AvgGlobalHoriz") 

all_years_df8 = all_years_df8.withColumnRenamed("<lambda>(Date AS Event)", "Event")
display(all_years_df8)
```
